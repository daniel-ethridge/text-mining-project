```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("../text-mining-project-data/clean/count-4856.csv")
labels = df.pop("labels")

enc = LabelEncoder()
enc_labels = enc.fit_transform(labels)
```

```python
trans = TfidfTransformer()
tfidf = trans.fit_transform(df).toarray()
tfidf_df = pd.DataFrame(tfidf, columns = trans.get_feature_names_out())
tfidf_df_lab = tfidf_df.copy()
tfidf_df_lab.insert(0, "label", enc_labels)
x_train, x_test, y_train, y_test = train_test_split(tfidf_df, enc_labels)
tfidf_df_lab.sample(10)
```
